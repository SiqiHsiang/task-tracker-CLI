```python
import json
import os
from datetime import datetime

# Add, Update, and Delete tasks
# Mark a task as in progress or done
# List all tasks
# List all tasks that are done
# List all tasks that are not done
# List all tasks that are in progress
```

```python
TASKS_FILE = "tasks.json"

#Load and save tasks

def load_tasks():
    if not os.path.exists(TASKS_FILE):
        return []
    with open(TASKS_FILE, 'r') as f:
        return json.load(f)

def save_tasks(tasks):
    with open(TASKS_FILE,'w') as f:
        json.dump(tasks, f, indent=2)
```

```python
#Add a task

def add_task(description):
   tasks = load_tasks()
   new_id = max([task['id'] for task in tasks], default=0) + 1
   created_at = datetime.now().isoformat()
   update_at = created_at
   task = {
      "id": new_id,
      "description": description,
      "status": "todo",
      "createdAt": created_at,
      "updatedAt": update_at
   }
   tasks.append(task)
   save_tasks(tasks)
   print(f"‚úÖ Task added successfully (ID: {new_id})")
```

```python
#Delete a task

def delete_task(task_id = None, description=None):

    tasks = load_tasks()

    if task_id is not None:    
        updated_tasks = [task for task in tasks if task["id"] != task_id]
        if len(updated_tasks) == tasks:
            print(f"‚ö†Ô∏è Task {task_id} not found.")
        else:
            save_tasks(updated_tasks)
            print(f"üóëÔ∏è Task {task_id} deleted.")
        return
    
    if description is not None:
        matches = [task for task in tasks if task["description"] == description]
        if len(matches) == 0:
            print(f"‚ö†Ô∏è Task {description} not found.")
            return
        elif len(matches) == 1:
            tasks.remove(matches[0])
            save_tasks(tasks)
            print(f"üóëÔ∏è Task {description} deleted.")
            return
        else:
            print(f"‚ö†Ô∏è Found multiple tasks with description '{description}'. Please specify the task ID.")
            for task in matches:
                print(f"ID: {task['id']} | Status: {task['status']} | Created: {task['createdAt']} | Updated: {task['updatedAt']}")
            return
    
    print("‚ö†Ô∏è Please provide either a task_id or a description to delete.")
```

```python
#Update a task

def update_task(task_id = None, new_description = None):
    tasks = load_tasks()
    for task in tasks:
        if task['id'] == task_id:
            task['description'] = new_description
            task["updatedAt"] = datetime.now().isoformat()
            save_tasks(tasks)
            print(f"‚úçÔ∏è Task {task_id} updated successfully.")
            return
        else:
            print(f"‚ö†Ô∏è Task {task_id} not found.")
```

```python
#Mark a state

def mark_task_status(task_id, new_status):
    valid_statuses = ["todo", "in-progress", "done"]
    if new_status not in valid_statuses:
        print(f"‚ö†Ô∏è Invalid status: '{new_status}'. Must be one of {valid_statuses}.")
        return
    
    tasks = load_tasks()
    for task in tasks:
        if task["id"] == task_id:
            task["status"] = new_status
            task["updatedAt"] = datetime.now().isoformat()
            save_tasks(tasks)
            print(f"üîñ Task {task_id} marked as '{new_status}'.")
            return
    
    print(f"‚ö†Ô∏è Task {task_id} not found.")
```

```python
#List tasks

def list_tasks(status=None):
    tasks = load_tasks()

    if status:
        valid_statuses = ["todo", "in-progress", "done"]
        if status not in valid_statuses:
            print(f"‚ö†Ô∏è Invalid status: '{status}'. Use one of {valid_statuses}.")
            return
        tasks = [task for task in tasks if task["status"] == status]
        print(f"üìã Tasks with status '{status}':\n")
    else:
        print("üìã All Tasks:\n")

    if not tasks:
        print("üì≠ No tasks found.")
        return

    for task in tasks:
        print(f"- ID: {task['id']}")
        print(f"  Description: {task['description']}")
        print(f"  Status: {task['status']}")
        print(f"  Created: {task['createdAt']}")
        print(f"  Updated: {task['updatedAt']}")
        print("-" * 30)
```
